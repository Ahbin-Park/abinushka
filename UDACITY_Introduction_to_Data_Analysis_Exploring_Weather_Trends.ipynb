```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
tem_moscow = pd.read_csv('/content/drive/MyDrive/tem_moscow.csv', index_col=0)
tem_johannesburg = pd.read_csv('/content/drive/MyDrive/tem_johannesburg.csv', index_col=0)
tem_seoul = pd.read_csv('/content/drive/MyDrive/tem_seoul.csv', index_col=0)
tem_world = pd.read_csv('/content/drive/MyDrive/tem_world.csv', index_col=0)
```

```python
sma_7d_tem_seoul = tem_seoul.rolling(window=7).mean()
```

```python
sma_7d_tem_world = tem_world.rolling(window=7).mean()
```

```python
sma_7d_tem_moscow = tem_moscow.rolling(window=7).mean()
```

```python
sma_7d_tem_johannesburg = tem_johannesburg.rolling(window=7).mean()
```

```python
sma_7d_tem_seoul.plot(kind= 'line')
```

```python
sma_7d_tem_world.plot(kind= 'line')
```

```python
import seaborn as sns
```

```python
sns.lineplot(x='year', y='avg_temp', data=sma_7d_tem_seoul, label = 'temperature trend of seoul')
sns.lineplot(x='year', y='avg_temp', data=sma_7d_tem_moscow, label = 'temperature trend of moscow')
sns.lineplot(x='year', y='avg_temp', data=sma_7d_tem_johannesburg, label = 'temperature trend of johannesburg')
sns.lineplot(x='year', y='avg_temp', data=sma_7d_tem_world, label = 'temperature trend of world')

plt.xlabel('year')
plt.ylabel('temperature')
plt.show()
```
